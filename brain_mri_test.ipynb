```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

dif_file = "med_data/lgg-mri-segmentation/kaggle_3m/TCGA_CS_5396_20010302/TCGA_CS_5396_20010302_1_mask.tif"

# Load the original .tif in grayscale
img = cv2.imread(tif_file, cv2.IMREAD_GRAYSCALE)
# Convert to RGB just for nicer colors in matplotlib
img_rgb = cv2.cvtColor(cv2.imread(tif_file), cv2.COLOR_BGR2RGB)

polys = tif_to_polygons(tif_file)

# Create a copy for visualizing contours
contoured_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
for poly in polys:
    cv2.drawContours(contoured_img, [poly], -1, (0,255,0), 2)

# Display side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(img_rgb)
axes[0].set_title("Original Image")
axes[1].imshow(contoured_img)
axes[1].set_title(f"Polygons (found {len(polys)})")
plt.show()
```
